```python
import pandas as pd

# Movie titles with their genres (labelled)
data = {
    "title": ["Love in Paris", "Blood Night", "Romantic Rain", "Haunted Forest", "Valentine Diary", "Ghost in the Mirror"],
    "label": ["Romantic", "Horror", "Romantic", "Horror", "Romantic", "Horror"]
}

df = pd.DataFrame(data)
df
```

```python
# Add new columns (features)
df['title_length'] = df['title'].apply(len)
df['word_count'] = df['title'].apply(lambda x: len(x.split()))
df
```

```python
df['label_encoded'] = df['label'].map({'Romantic': 0, 'Horror': 1})
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Features and labels
X = df[['title_length', 'word_count']]
y = df['label_encoded']

# Split into train and test (we’ll use all data to keep it easy)
model = LogisticRegression()
model.fit(X, y)
```

```python
# Try predicting a new title
new_title = "The Dark Knight"
title_length = len(new_title)
word_count = len(new_title.split())

# Predict
prediction = model.predict([[title_length, word_count]])
print("Predicted Genre:", "Horror" if prediction[0] == 1 else "Romantic")
```
